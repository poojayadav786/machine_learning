## Mutivariate Analysis
Multivariate Data Analysis refers to any statistical technique used to analyze data that arises from more than one variable. This essentially models reality where each situation, product, or decision involves more than a single variable.The techniques are especially valuable when working with correlated variables.

```python
grid=sns.FacetGrid(df,col='quality',col_wrap=2)
grid.map(plt.scatter,'alcohol','density',alpha=0.3)
plt.subplots_adjust(top=0.9)
grid.fig.suptitle('Alcohol v/s Density in terms of  Quality')
plt.show()
```

### Inferences from the above Alcohol v/s Density in terms  of Quality plot.
This Facegrid highlights the differences in alcohol content and density in the top ranked wines and the lowest ranked wines. Generally, high quality wines have higher alcohol content and lower density.
However, there are many observations which do not follow this general trend. Because alcohol and density are highly correlated,and also Alcohol and quality also highly correlated so we can't notice big difference.

```python
grid=sns.FacetGrid(df,col='quality',col_wrap=2)
grid.map(plt.scatter,'fixed acidity','alcohol',alpha=0.3)
plt.subplots_adjust(top=0.9)
grid.fig.suptitle('Alcohol v/s fixed acidity in terms of Quality ')
plt.show()
```

### Inferences from the above Alcohol v/s fixed acidity in terms  of Quality plot.
There is a lot of variance in this plot, much more than in the last one. This is likely because there is low correlation between alcohol and  fixed acidity(-1.2). However, it is clear from this plot that the highest ranked wines have higher alcohol and lower acidity. However,this help in follow the general trend. 

```python
grid=sns.FacetGrid(df,col='quality',col_wrap=2)
grid.map(plt.scatter,'free sulfur dioxide','alcohol',alpha=0.3)
plt.subplots_adjust(top=0.9)
grid.fig.suptitle('Alcohol v/s free sulfur dioxide in terms of Quality ')
plt.show()
```

### Inferences from the above Alcohol v/s free sulfur dioxide in terms of Quality plot
This Facetgrid demonstrates again that higher quality wines tend to have higher alcohol content. It explains less of the variation in the free sulfur dioxide variable between quality levels, except that higher quality wines have less varied free sulfur dioxide levels.
